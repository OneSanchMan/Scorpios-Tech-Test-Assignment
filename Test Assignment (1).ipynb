```python
from sklearn.metrics import mean_absolute_error
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
from sklearn.preprocessing import LabelEncoder

# Create target object and call it y
y = churn_users.churn
# Create X
features = ['os_version', 'latest_version', 'user_properties.best speed', 
            'user_properties.current speed',
            'completed game', 'completed sublevel', 'session_end',
            'session_start', 'clicked tab', 'Started Test',
            'Viewed subscription offer', 'Closed subscription offer',
            'Finished test', 'completed name', 'completed phone',
            'completed genre page', 'started sublevel', 'completed level',
            'closed sublevel', 'started game', 'Clicked subscribe',
            'Sent Campaign Name', 'Started Free Trial', 'changed font size',
            'changed genre', 'closed subscription', 'Subscribed',
            'closed subsciption', 'Renewed subscription', 'Cancelled free trial',
            'subscribed', 'Started trial', 'Cancelled trial',
            'Cancelled subscription', 'Subscribed after trial', 'test just']
X_train = churn_users[features]
X = active_users[features]

X_train = pd.DataFrame(X_train).fillna(value=0)
X = pd.DataFrame(X).fillna(value=0)
X_train
```

```python
object_cols = [col for col in X_train.columns if X_train[col].dtype == "object"]

# Columns that can be safely label encoded
good_label_cols = [col for col in object_cols if 
                   set(X_train[col]) == set(X[col])]
bad_label_cols = list(set(object_cols)-set(good_label_cols))

label_X_train = X_train.drop(bad_label_cols, axis=1)
label_X = X.drop(bad_label_cols, axis=1)

# Apply label encoder
label_encoder = LabelEncoder()
for col in set(good_label_cols):
    label_X_train[col] = label_encoder.fit_transform(X_train[col])
    label_X[col] = label_encoder.transform(X[col])
```

```python
label_X
```

```python
model = RandomForestRegressor(random_state=1)

model.fit(label_X_train, y)
```

```python
users_X = active_users[features].fillna(method='bfill')
users_X
```

```python
y_predict = model.predict(label_X)
list(active_users['user_properties.amplitude_id'])
```

```python
data = {'amplitude_id':list(active_users['user_properties.amplitude_id']),
        'segment':[0] * len(y_predict)}

df = pd.DataFrame(data)
for i in range(len(y_predict)):
    if y_predict[i] >= 0.75:
        df.segment[i] = 2
    elif y_predict[i] <= 0.25:
        df.segment[i] = 1
    else:
        df.segment[i] = 3
df
```
